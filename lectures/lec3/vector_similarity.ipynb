```python
import numpy as np
import matplotlib.pyplot as plt
from sktime.distances import dot_distance

x = [0, 1, 2, 3, 4]
x1 = [3, 3, 5, 3, 2]
x2 = [4, 2, 4, 2, 3]


def plot_similar_but_different_curves():
    plt.figure(figsize=(8, 6))
    plt.plot(x, x1, label='Curve 1', color='b')
    plt.plot(x, x2, label='Curve 2', color='purple')
    plt.xticks(range(int(min(x)), int(max(x)) + 1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()


x1 = np.asarray(x1)
x2 = np.asarray(x2)

# Dot product
dot_product = np.dot(x1, x2)
# or
# dot_product = x1 @ x2

# Cosine similarity
norm_product = np.linalg.norm(x1) * np.linalg.norm(x2)
cosine_similarity = 0.0 if norm_product == 0 else dot_product / norm_product

# sktime dot_distance = 1 - cosine_similarity
dot_dist = dot_distance(x1, x2)

print("Dot product:", dot_product)
print("Cosine similarity:", cosine_similarity)
print("dot_distance (1 - cosine_similarity):", dot_dist)
plot_similar_but_different_curves()

```
