```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sktime.distances import dcor1_distance, dcor2_distance, pearson_correlation

# Generate time series with strong positive correlation
t = np.linspace(0, 10, 50)
x1 = t * 0.1 + np.random.normal(0, 0.1, 50)
x2 = t * 1 + np.random.normal(0, 1, 50)

# Calculate correlation
pearson_corr = pearson_correlation(x1, x2)

# First form of dCOR (dCOR1)
dcor1 = dcor1_distance(x1, x2)

# Second form of dCOR (dCOR2), using a chosen beta value
beta = 1.0
dcor2 = dcor2_distance(x1, x2, beta=beta)

# Plot
sns.set()
plt.figure(figsize=(15, 4))

# Plot x1
plt.subplot(121)
plt.title('x1')
plt.plot(t, x1)
plt.grid(True)

# Plot x2
plt.subplot(122)
plt.title('x2')
plt.plot(t, x2)
plt.grid(True)

# Add Pearson, dCOR1, and dCOR2 to the plot title
plt.suptitle(f"Pearson Correlation: {pearson_corr:.3f}, dCOR1: {dcor1:.3f}, dCOR2: {dcor2:.3f}")

plt.tight_layout()
plt.show()
```
