```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
# Frequency and sampling rate
f = 10  # frequency
Fs = 100  # sampling rate
t = np.arange(0, 1, 1/Fs)
# Sine function
y = np.sin(2 * np.pi * f * t)
# Perform Fourier transform using scipy
y_fft = fftpack.fft(y)
# Plot data
n = np.size(t)
fr = np.linspace(0, Fs/2, n//2)  # Correct the frequency array calculation
y_m = 2/n * np.abs(y_fft[:np.size(fr)])  # Take the first half of the FFT results
# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
# Plot time domain
ax[0].plot(t, y)
# Plot frequency domain
ax[1].stem(fr, y_m)
plt.show()
```
