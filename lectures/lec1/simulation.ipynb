```python
import numpy as np
import matplotlib.pyplot as plt

# Define time series
time = np.arange(0, 96, 1)

# 1. Linear Trend
linear_trend = 0.5 * time

# 2. Curved Trend (Quadratic)
curved_trend = 0.005 * time**2

# 3. Trend with Sinusoidal Period
trend_with_drift = np.sin(0.1 * time) * np.mean(curved_trend) / 2
composite_2 = curved_trend + trend_with_drift

# 4. More Complex Periodic Component
avg_aug = np.mean(linear_trend) / 4
complex_periodic_component = avg_aug * np.sin(0.5 * time) + avg_aug * np.sin(0.7 * time)
composite_3 = linear_trend + complex_periodic_component

# 5. Random Noise
noise = avg_aug * np.random.normal(0, 1, len(time))
composite_4 = composite_3 + noise

# Plotting all 5 charts
fig, axes = plt.subplots(5, 1, figsize=(10, 20), constrained_layout=True)

# Plot 1: Linear Trend vs. Curved Trend
axes[0].plot(time, linear_trend, 'b--', label='Linear Trend')
axes[0].plot(time, curved_trend, color='purple', label='Curved Trend')
axes[0].set_title('Linear Trend vs. Curved Trend')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Value')
axes[0].legend()
axes[0].grid()

# Plot 2: Pure Sinusoidal Period and Composite with Quadratic Trend
axes[1].plot(time, trend_with_drift, color='purple', label='Pure Sin Period')
axes[1].set_title('Pure Sin Period')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
axes[1].legend()
axes[1].grid()

axes[2].plot(time, composite_2, color='purple', label='Trend + Sin')
axes[2].plot(time, curved_trend, 'b--', label='Trend (Step 2)')
axes[2].set_title('Trend with Period')
axes[2].set_xlabel('Time')
axes[2].set_ylabel('Value')
axes[2].legend()
axes[2].grid()

# Plot 3: Complex Periodic Component
axes[3].plot(time, complex_periodic_component, color='purple', label='Pure Period')
axes[3].set_title('Pure Periodic Component')
axes[3].set_xlabel('Time')
axes[3].set_ylabel('Value')
axes[3].legend()
axes[3].grid()

axes[4].plot(time, composite_4, color='purple', label='Trend + Period + Noise')
axes[4].plot(time, composite_3, 'b--', label='Trend + Period (Step 3)')
axes[4].set_title('Trend with Complex Periodic Component and Noise')
axes[4].set_xlabel('Time')
axes[4].set_ylabel('Value')
axes[4].legend()
axes[4].grid()

# Display all plots
plt.show()
```
